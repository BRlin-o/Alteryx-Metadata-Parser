```python
import csv
import sys
from pathlib import Path
from shutil import copyfile
from Node import NodeElement
import xml.etree.ElementTree as ET

# Input file
file_path = Path("./sample_alteryx_workflow.yxmd")

# 檢查文件是否有擴展名
if not file_path.suffix:
    print('Error: Input file must have an extension')

# 檢查文件擴展名是否合適
if file_path.suffix not in ['.xml', '.yxmd']:
    print('Error: Input file must be .xml or .yxmd')

if file_path.suffix == '.yxmd':
    xml_path = file_path.with_suffix('.xml')
    copyfile(file_path, xml_path)
    tree = ET.parse(xml_path)
else:
    tree = ET.parse(file_path)

output_folder = Path("./output")
if not output_folder.exists():
    output_folder.mkdir()

# Output file
output_path = output_folder / "output.csv"
print("Input file: " + str(file_path))
print("Output file: " + str(output_path))
```

```python
root = tree.getroot()

lst = []
for x in root.iter('Node'):
    node = NodeElement(x)
    lst.append(node.data)

lst
```

```python
keys = lst[0].keys()
with open(output_path, 'w') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(lst)
```
